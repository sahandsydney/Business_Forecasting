```python
import requests
import json

# Set the GDELT API endpoint
api_url = "https://api.gdeltproject.org/api/v2/summary/summary"

# Set the search parameters
search_terms = "Adani"
start_date = "2004-01-01"
end_date = "2022-12-31"
format = "json"

# Set the query parameters
params = {
    "query": search_terms,
    "startdatetime": start_date,
    "enddatetime": end_date,
    "format": format,
}

# Send the request to the GDELT API
response = requests.get(api_url, params=params)

# Parse the JSON response
data = response.json()

# Process the news articles
articles = data["articles"]

# Save the articles to a CSV file
csv_filename = "adani_news.csv"
with open(csv_filename, "w") as file:
    for article in articles:
        title = article["title"]
        date = article["date"]
        source = article["source"]
        url = article["url"]
        file.write(f"{date},{title},{source},{url}\n")
```

```python
import csv
from newspaper import Article
from datetime import datetime
from datetime import datetime, timedelta


# Set the URL of the news source
news_url = 'https://timesofindia.indiatimes.com/topic/adani?dateFilter=1088965800000,165704579900'

# Create a list to store the extracted articles
articles = []

# Iterate over the desired date range
start_date = datetime(2022, 12, 1)
end_date = datetime(2022, 12, 31)
date_range = [start_date + timedelta(days=x) for x in range((end_date - start_date).days + 1)]

for date in date_range:
    # Construct the URL for the specific date
    formatted_date = date.strftime('%Y-%m-%d')
    url = f'{news_url}/{formatted_date}'

    # Create an Article object and download/parse the article
    article = Article(url)
    article.download()
    article.parse()

    # Extract relevant information from the article
    title = article.title
    publish_date = article.publish_date.strftime('%Y-%m-%d')
    text = article.text

    # Create a dictionary to store the article data
    article_data = {
        'Title': title,
        'Publish Date': publish_date,
        'Text': text
    }

    # Append the article data to the list
    articles.append(article_data)

# Define the CSV file path
csv_file = 'adani_news.csv'

# Write the articles to the CSV file
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['Title', 'Publish Date', 'Text'])
    writer.writeheader()
    writer.writerows(articles)

print(f'News articles extracted and saved to {csv_file}.')
```

```python
import csv
from newspaper import Article
from datetime import datetime, timedelta

# Set the URL of the news source (Times of India)
news_url = 'https://timesofindia.indiatimes.com/topic/adani?dateFilter=1088965800000,165704579900'

# Create a list to store the extracted articles
articles = []

# Iterate over the desired date range
start_date = datetime(2022, 12, 1)
end_date = datetime(2022, 12, 31)
date_range = [start_date + timedelta(days=x) for x in range((end_date - start_date).days + 1)]

# Define the CSV file path
csv_file = 'adani_news_test.csv'

# Write the articles to the CSV file
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Publish Date', 'Text'])

    for date in date_range:
        # Construct the URL for the specific date
        formatted_date = date.strftime('%Y/%m/%d')
        url = f'{news_url}/{formatted_date}'

        # Create an Article object and download/parse the article
        article = Article(url)
        article.download()
        article.parse()

        # Extract relevant information from the article
        title = article.title
        publish_date = article.publish_date.strftime('%Y-%m-%d')
        text = article.text

        # Write the article data to the CSV file
        writer.writerow([title, publish_date, text])

        # Append the article data to the list
        articles.append({
            'Title': title,
            'Publish Date': publish_date,
            'Text': text
        })

print(f'News articles extracted and saved to {csv_file}.')
```
